```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer to convert texts to vectors
vectorizer = CountVectorizer().fit_transform(data)

# Calculate similarity
similarity = cosine_similarity(vectorizer)

# Create a dictionary to store the count of filtered data for each keyword
filtered_data_count2 = {keyword: 0 for keyword in keywords2}

#Define the threshold
threshold = 0.1

# Iterate over texts and their similarities
for i, text in enumerate(data):
    for j, keyword in enumerate(keywords2):
        if keyword in text and similarity[i][j] >= threshold:
            filtered_data_count2[keyword] += 1

print(filtered_data_count2)
```

```python
total_filtered_data = sum(filtered_data_count2.values())

print(total_filtered_data)
```

# Jaccard

```python
from nltk.metrics.distance import jaccard_distance
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
import nltk
from nltk.metrics.distance import jaccard_distance
from nltk.stem import SnowballStemmer
from nltk.stem.snowball import SnowballStemmer

threshold = 0.1

filtered_jacc = {}

# Create stemmer object
factory = StemmerFactory()
stemmer = factory.create_stemmer()

for keyword in keywords:
    filtered_jacc[keyword] = []
    for item in data:
        # Stem the item and keyword
        stemmed_item = stemmer.stem(item)
        stemmed_keyword = stemmer.stem(keyword)
        # Calculate similarity
        similarity = 1 - jaccard_distance(set(nltk.word_tokenize(stemmed_item)), set(nltk.word_tokenize(stemmed_keyword)))
        if similarity >= threshold:
            filtered_jacc[keyword].append(item)

total_filtered_data = 0
for keyword, data in filtered_jacc.items():
    print(f"Filtered data for keyword '{keyword}': {data}")
    print(f"Count of filtered data for keyword '{keyword}': {len(data)}")
    total_filtered_data += len(data)

print(f"Total count of filtered data: {total_filtered_data}")

```
